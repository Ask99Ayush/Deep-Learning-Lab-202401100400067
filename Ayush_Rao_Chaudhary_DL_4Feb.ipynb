```python
import tensorflow as tf
print(f"TensorFlow Version: {tf.__version__}")
```

```python
scalar_tensor = tf.constant(7)
print("Scalar Tensor:", scalar_tensor)


vector_tensor = tf.constant([1, 2, 3])
print("Vector Tensor:", vector_tensor)


matrix_tensor = tf.constant([[1, 2], [3, 4]])
print("Matrix Tensor:\n", matrix_tensor)
```

```python
tensor_a = tf.constant([[10., 20.], [30., 40.]])
tensor_b = tf.constant([[ 1.,  2.], [ 3.,  4.]])

print("Tensor A:\n", tensor_a)
print("Tensor B:\n", tensor_b)


addition_result = tensor_a + tensor_b
print("\nAddition (A + B):\n", addition_result)


subtraction_result = tensor_a - tensor_b
print("\nSubtraction (A - B):\n", subtraction_result)


multiplication_result = tensor_a * tensor_b
print("\nElement-wise Multiplication (A * B):\n", multiplication_result)


division_result = tensor_a / tensor_b
print("\nElement-wise Division (A / B):\n", division_result)
```

```python
integer_tensor = tf.constant([1, 2, 3, 4, 5])
print("Integer Tensor:", integer_tensor)
print("  Data Type:", integer_tensor.dtype)
print("  Shape:", integer_tensor.shape)


float_tensor = tf.constant([1.0, 2.5, 3.8, 4.1], dtype=tf.float32)
print("\nFloat Tensor:", float_tensor)
print("  Data Type:", float_tensor.dtype)
print("  Shape:", float_tensor.shape)


matrix_tensor = tf.constant([[10, 20], [30, 40], [50, 60]])
print("\nMatrix Tensor:", matrix_tensor)
print("  Data Type:", matrix_tensor.dtype)
print("  Shape:", matrix_tensor.shape)
```

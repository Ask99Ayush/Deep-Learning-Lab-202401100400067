```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Iris.csv")
df.shape
```

```python
df.info()
df.describe(include="all")
```

```python
df.isnull().sum()
df.duplicated().sum()

df = df.drop_duplicates()
df = df.ffill()
```

```python
df["Id_normalized"] = (df["Id"] - df["Id"].min()) / (df["Id"].max() - df["Id"].min())
df.head()
```

```python
plt.figure()
df["Id"].plot()
plt.title("Numeric Column Trend")
plt.show()
```
